# Clean Text

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

stopwords = set(stopwords.words('english'))
```

```python
def clean(text):
    text=text.lower()

    text=re.sub(r'[^a-zA-z0-9\s]','',text)

    text=word_tokenize(text)

    text=[w for w in text if w not in stopwords]
    return " ".join(text)
```

```python
text="My name ... @ Priya Singh00012"
clean(text)
```

```python
df['TITLE_NEW']=df['TITLE'].apply(clean)
```

```python
from textblob import TextBlob
from collections import Counter
```

```python
from textblob import TextBlob
from collections import Counter

def get_keywords(text):
    blob = TextBlob(text)

    # Extract only the desired POS tags
    keywords = [word for word, tag in blob.tags if tag in (
        'NN', 'NNS', 'NNP', 'NNPS',  # Nouns
        'JJ', 'JJR', 'JJS',          # Adjectives
        'RB', 'RBR', 'RBS'           # Adverbs
    )]

    keyword_counts = Counter(keywords)
    most_common_keywords = keyword_counts.most_common(5)
    return most_common_keywords
```

```python
df['keywords']=df['TITLE_NEW'].apply(get_keywords)
```
